```python
#In the excel, I deleted some unnecessary rows, transposed my original data and switched one column from a percentage to a 
#number.
#Import libraries

import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

#%matplotlib.inline
#matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots

#read in the data
df = pd.read_csv(r"C:\Users\eknag\Downloads\Micron corelation data.csv")
df = df.drop(df.columns[7:], axis=1)
df = df.dropna()
```

```python
df
```

```python
plt.scatter(x=df['Revenue'], y=df['Stock price'])
plt.title('Revenue vs Stock Price')
plt.xlabel('Revenue')
plt.ylabel('Stock Price')
plt.show()
```

```python
sns.regplot(x='Revenue', y="Stock price", data=df, scatter_kws={'color':'green'}, line_kws={'color':'orange'})
```

```python
df.corr(numeric_only=True, method='pearson')
```

```python
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix)
plt.title('Correlation Matrix')
plt.show()
```

```python
corr_pairs = correlation_matrix.unstack()
sorted_pairs = corr_pairs.sort_values()
high_corr = sorted_pairs[sorted_pairs > 0.5]
high_corr.drop_duplicates()
```
